```python
import pandas as pd

# Misal dataframe kamu bernama df
# Tentukan kolom numerik yang akan dicek
kolom_numerik = ['Sales', 'Quantity', 'Discount', 'Profit', 'Shipping Cost']

# Loop tiap kolom untuk cek outlier
for kolom in kolom_numerik:
    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR

    outlier = df[(df[kolom] < batas_bawah) | (df[kolom] > batas_atas)]
    print(f"Kolom: {kolom}")
    print(f" Jumlah outlier: {outlier.shape[0]}")
    print(f" Batas bawah: {batas_bawah:.2f}, Batas atas: {batas_atas:.2f}")
    print("-"*50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
df[kolom_numerik].boxplot()
plt.title("Boxplot untuk Deteksi Outlier")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pisahkan data
# Recreate the boolean mask for outliers
outlier_mask = (df[kolom_numerik] < batas_bawah) | (df[kolom_numerik] > batas_atas)
# Combine the masks across all numeric columns
combined_outlier_mask = outlier_mask.any(axis=1)

df_outlier = df[combined_outlier_mask]
df_bersih = df[~combined_outlier_mask]

# Hasil
print(f"Jumlah total data: {len(df)}")
print(f"Jumlah data outlier: {len(df_outlier)}")
print(f"Jumlah data bersih: {len(df_bersih)}")
```

4. Berapa lama waktu yang diperlukan untuk memproses dan mempersiapkan pesanan sebelum pengiriman (order processing time)?

```python
# Ubah kolom 'Order Date' dan 'Ship Date' menjadi tipe data datetime
df_bersih['Order Date'] = pd.to_datetime(df_bersih['Order Date'])
df_bersih['Ship Date'] = pd.to_datetime(df_bersih['Ship Date'])

# Hitung selisih waktu antara 'Order Date' dan 'Ship Date'
df_bersih['Processing Time'] = (df_bersih['Ship Date'] - df_bersih['Order Date']).dt.days

# Hitung rata-rata processing time
rata_rata = df_bersih['Processing Time'].mean()
(f"Rata-rata waktu pemrosesan pesanan: {rata_rata:.2f} hari")
```

# **Analisis Trend**

1. Bagaimana trend total penjualan dan volume penjualan setiap bulannya?


```python
# Ubah tipe data kolom 'Order Date' menjadi datetime
df_bersih['Order Date'] = pd.to_datetime(df_bersih['Order Date'])

# Ekstrak bulan dari kolom 'Order Date'
df_bersih['Month'] = df_bersih['Order Date'].dt.month

# Kelompokkan data berdasarkan bulan dan hitung total penjualan dan volume penjualan
monthly_sales = df_bersih.groupby('Month')['Sales'].sum()
monthly_quantity = df_bersih.groupby('Month')['Quantity'].sum()

# Buat visualisasi tren menggunakan line chart
plt.figure(figsize=(12, 6))

plt.plot(monthly_sales.index, monthly_sales.values, label='Total Penjualan')
plt.plot(monthly_quantity.index, monthly_quantity.values, label='Volume Penjualan')

plt.title('Tren Total Penjualan dan Volume Penjualan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Nilai')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)

plt.show()
```

**Kesimpulan : Trend total penjualan jika dilihat berdasarkan volume penjualan dan total penjualan yaitu menunjukkan peningkatan setiap bulannya, dengan penjualan tertinggi terjadi di akhir tahun yaitu bulan desember**

2. Bagaimana trendnya jika dilihat berdasarkan kategori produk.

```python
# Kelompokkan data dan unstack kategori
monthly_sales = df_bersih.groupby([pd.Grouper(key='Order Date', freq='M'), 'Category'])['Sales'].sum().unstack()

# Ubah nama index menjadi 'Month'
monthly_sales.index.name = 'Month'

# Visualisasi menggunakan line chart
monthly_sales.plot(kind='line', figsize=(12, 6))
plt.title('Trend Penjualan per Bulan Berdasarkan Kategori Produk')
plt.xlabel('Bulan')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan : Trend total penjualan jika dilihat berdasarkan kategori produk setiap bulannya yaitu penjualan setiap bulannya berbeda-beda untuk setiap kategori produk, namun cenderung mengalami kenaikkan**
