```python
outliers_dict = {}      # simpan outliers per kolom
clean_dict = {}         # simpan data tanpa outliers per kolom

for col in numerical_cols:
  print(f'-----{col}-----')
  Q1 = dfB_clean[col].quantile(0.25)
  Q2 = dfB_clean[col].quantile(0.5)
  Q3 = dfB_clean[col].quantile(0.75)
  IQR = Q3-Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = dfB_clean[(dfB_clean[col] < lower_bound) | (dfB_clean[col] > upper_bound)]
  clean = dfB_clean[(dfB_clean[col] >= lower_bound) & (dfB_clean[col] <= upper_bound)]

  outliers_dict[col] = outliers
  clean_dict[col] = clean

  print(f"Kuartil 1: {Q1}, Kuartil 2: {Q2}, Kuartil 3: {Q3}")
  print(f"Max: {dfB_clean[col].max()},  Min: {dfB_clean[col].min()}")
  print(f"Banyaknya Outliers: {len(outliers)}")

  plt.figure(figsize = (10,4))
  plt.subplot(1,2,1)
  dfB_clean[col].hist(bins = 20)
  plt.title(f'Distribution of {col}')

  plt.subplot(1,2,2)
  dfB_clean.boxplot(column = col)
  plt.title(f'Boxplot of {col}')
  plt.show()
```

## 6 Jika RPPI ingin melakukan investasi (max) Rp. 25 Milyar minggu besok, apakah dari data anda dapat memberikan rekomendasi rumah mana saja yang akan berpotensi menghasilkan keuntungan bagi perusahaan?

```python
# ---------- 2) Metrik dasar: harga per m2 ----------
df_all['harga_per_m2_bangunan'] = df_all['harga'] / df_all['LB']
df_all['harga_per_m2_tanah'] = df_all['harga'] / df_all['LT']
```

```python
kota_stats = df_all.groupby('Kota').agg(
    median_harga=('harga','median'),
    median_ppm2_bangunan=('harga_per_m2_bangunan','median'),
    count_listings=('harga','count')
).reset_index()

df = df_all.merge(kota_stats, on='Kota', how='left')

# ---------- 4) Tandai "undervalued" relatif ke median kota ----------
# kriteria undervalued (bisa disesuaikan):
# - harga < median_harga_kota
# - harga_per_m2_bangunan < median_ppm2_bangunan_kota
df['undervalued_price'] = df['harga'] < df['median_harga']
df['undervalued_ppm2'] = df['harga_per_m2_bangunan'] < df['median_ppm2_bangunan']

# Buat skor sederhana (composite) untuk peringkat peluang:
# skor basic: undervalued_price (1) + undervalued_ppm2 (1) + (relatif murah dari median) * faktor
df['relatif_diskon'] = (df['median_harga'] - df['harga']) / df['median_harga']  # proporsi "diskon" relatif
df['relatif_diskon'] = df['relatif_diskon'].clip(lower=0)  # hanya positif = diskon

# normalisasi relatif_diskon supaya scale wajar
max_disc = df['relatif_diskon'].max() if df['relatif_diskon'].max()>0 else 1
df['relatif_diskon_norm'] = df['relatif_diskon'] / max_disc

# composite score (0..~3) â€” atur bobot sesuai preferensi
w_price = 0.4
w_ppm2  = 0.3
w_disc  = 0.3
df['score'] = (w_price * df['undervalued_price'].astype(int) +
               w_ppm2  * df['undervalued_ppm2'].astype(int) +
               w_disc  * df['relatif_diskon_norm'])

# ---------- 5) Filter properti yang bisa dibeli (harga <= budget) ----------
budget = 25_000_000_000  # 25 M
df_affordable = df[df['harga'] <= budget].copy()

# bila ingin hanya properti non-outlier global, bisa filter di sini (opsional)

# ---------- 6) Ranking & Portofolio greedy: pilih sampai anggaran terpenuhi ----------
# Urutkan: prioritas score tinggi, lalu harga rendah (untuk beli lebih banyak unit)
df_affordable = df_affordable.sort_values(by=['score','harga'], ascending=[False, True]).reset_index(drop=True)

# greedy pick
selected = []
cumsum = 0
for idx, row in df_affordable.iterrows():
    if cumsum + row['harga'] <= budget:
        selected.append(idx)
        cumsum += row['harga']

portfolio = df_affordable.loc[selected].copy()
portfolio['cumulative_cost'] = portfolio['harga'].cumsum()

# ---------- 7) Output ringkasan ----------
summary_portfolio = portfolio[['Kota','harga','LB','KT','KM','harga_per_m2_bangunan','score','cumulative_cost']]

# statistik pembelian per kota di portfolio
portfolio_city = portfolio.groupby('Kota').agg(
    units=('harga','count'),
    total_cost=('harga','sum'),
    avg_price=('harga','mean')
).reset_index()

# ---------- 8) Tampilan hasil ----------
print("Jumlah listing affordable:", len(df_affordable))
print("Jumlah unit terpilih dalam portofolio (greedy):", len(portfolio))
print("Total biaya portofolio:", portfolio['harga'].sum())
print("\nTop 10 kandidat (by score):")
print(df_affordable[['Kota','harga','LB','harga_per_m2_bangunan','score']].head(10))
print("\nPortofolio terpilih:")
print(summary_portfolio)
print("\nRingkasan per kota dalam portofolio:")
print(portfolio_city)
```

Dari hasil analisis menunjukkan bahwa kota Bogor, Depok, Bekasi menjadi kandidat utama untuk investasi karena:
- harga rata-rata kisaran 1-1.3 M cocok untuk bulk investment
- Potensi likuiditas tinggi (lebih mudah dijual kembali)
